```python
# Importing necessary libraries for data manipulation and visualization

# pandas: a powerful data manipulation library in Python
import pandas as pd

# numpy: a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np

# matplotlib: a plotting library for creating static, animated, and interactive visualizations in Python
import matplotlib.pyplot as plt

# seaborn: a Python data visualization library based on matplotlib, provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
# Loading the dataset
# pandas' read_csv function is used to read the csv file into a DataFrame

data = pd.read_csv("housing.csv")
```

```python
# Displaying the first few rows of the DataFrame
# This is useful to get a quick overview of the data's structure, columns, and values

data
```

```python
# Using the info() function to get a summary of the DataFrame
# This includes the number of entries, the number and names of the columns, the data type of each column, and the number of non-null entries in each column

data.info()
```

```python
# Removing missing values from the DataFrame
# The dropna() function removes missing values (NaN)
# The 'inplace=True' argument means that the changes are made directly to the original DataFrame

data.dropna(inplace=True)
```

```python
# Displaying the summary of the DataFrame again using the info() function
# This is to confirm that the missing values have been successfully removed

data.info()
```

```python
# Importing the train_test_split function from sklearn's model_selection module
# This function is used to split the dataset into training and testing sets

from sklearn.model_selection import train_test_split

# Separating the features (x) from the target variable (y)
# The target variable is 'median_house_value', which we're trying to predict
# The features are all the other columns in the DataFrame

x = data.drop(['median_house_value'], axis=1) 

y = data['median_house_value']
```

```python
# Splitting the features and target variable into training and testing sets
# The test_size argument specifies that 20% of the data will be used for the test set, and the rest will be used for the training set

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Joining the features (x_train) and the target variable (y_train) to form the complete training dataset
# The join() function is used to combine two dataframes based on their index

train_data = x_train.join(y_train)
```

```python
# Displaying the training dataset
# This is useful to confirm that the join operation in the previous cell was successful

train_data
```

```python
# Plotting histograms of all the columns in the training dataset
# This is useful to understand the distribution of values in each column
# The figsize argument specifies the size of the plot

train_data.hist(figsize=(15, 8))
```

```python
# Creating a heatmap to visualize the correlation between different features in the training dataset
# The heatmap is created using seaborn's heatmap function
# The correlation matrix is calculated using the corr() function
# The annot=True argument means that the correlation coefficients will be displayed on the heatmap
# The cmap argument specifies the color map to use

plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu") 
```

```python
# Applying a log transformation to the 'total_rooms', 'total_bedrooms', 'population', and 'households' columns
# The log transformation is a common technique used to handle skewed data
# Adding 1 before taking the log ensures that the values are defined (since log(0) is undefined)


train_data['total_rooms']= np.log(train_data['total_rooms'] + 1) 
train_data['total_bedrooms']= np.log(train_data['total_bedrooms'] + 1) 
train_data['population']= np.log(train_data['population'] + 1) 
train_data['households']= np.log(train_data['households'] + 1) 
```

```python
# Plotting histograms of all the columns in the training dataset again
# This is to visualize the effect of the log transformation on the distributions

train_data.hist(figsize=(15, 8))
```

```python
# Converting the 'ocean_proximity' column into dummy variables
# This is done because machine learning algorithms generally work better with numerical input
# The get_dummies function is used to create dummy variables
# The original 'ocean_proximity' column is dropped from the dataframe

train_data= train_data.join(pd.get_dummies( train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

```python
# Creating a heatmap to visualize the correlation between different features in the training dataset again
# This is done after the 'ocean_proximity' column has been converted into dummy variables
# The heatmap helps understand the relationships between the different features, including the newly created dummy variables

plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu") 
```

```python
# Creating a scatter plot of latitude and longitude to visualize the geographical distribution of house values
# The color of the points (hue) represents the median house value
# This plot can help identify patterns related to location, such as higher house values near the coast

plt.figure(figsize=(15, 8))
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
# Creating new features that might be relevant for predicting house values
# 'bedroom_ratio' is the ratio of total bedrooms to total rooms
# This could be a relevant feature because a house with more bedrooms might be more valuable

train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
# 'household_rooms' is the ratio of total rooms to households
# This could be a relevant feature because a house with more rooms per household might be more valuable

train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
# Creating a heatmap to visualize the correlation between different features in the training dataset again
# This is done after the new features 'bedroom_ratio' and 'household_rooms' have been added
# The heatmap helps understand the relationships between the different features, including the newly created ones

plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu") 
```

```python
# Importing the necessary modules for linear regression and data scaling
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Creating a StandardScaler object to scale the data
scaler = StandardScaler()

# Separating the features and target variable in the training dataset
# Also applying the scaling to the features
x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
x_train_s = scaler.fit_transform(x_train)

# Creating a LinearRegression object and fitting it to the training data
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Preparing the testing data
# This includes joining the features and target variable, applying the log 
# - transformation to certain columns, creating dummy variables, and creating the new features

test_data = x_test.join(y_test)

test_data['total_rooms']= np.log(test_data['total_rooms'] + 1) 
test_data['total_bedrooms']= np.log(test_data['total_bedrooms'] + 1) 
test_data['population']= np.log(test_data['population'] + 1) 
test_data['households']= np.log(test_data['households'] + 1) 

test_data= test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

# Creating the new features 'bedroom_ratio' and 'household_rooms'
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms'] 
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

```

```python
# Separating the features and target variable in the testing dataset

x_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
# Applying the same scaling to the features in the testing dataset
# It's important to use the same scaling for the training and testing data to ensure consistent results

x_test_s = scaler.transform(x_test)
```

```python
# Calculating the R^2 score of the linear regression model on the testing data
# The score method of a LinearRegression object returns the R^2 score, which measures the proportion of the variance in the target 
# - variable that can be predicted from the features

reg.score(x_test, y_test)
```

```python
# Importing the RandomForestRegressor class for hyperparameter tuning
from sklearn.ensemble import RandomForestRegressor

# Creating a RandomForestRegressor object and fitting it to the scaled training data
# Random forests are a type of ensemble learning method that can improve the predictive accuracy and control over-fitting

forest = RandomForestRegressor()
forest.fit(x_train_s, y_train)
```

```python
# Calculating the R^2 score of the random forest model on the scaled testing data
# The score method of a RandomForestRegressor object returns the R^2 score

forest.score(x_test_s, y_test)
```

```python
# Importing the GridSearchCV class for hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Creating a new RandomForestRegressor object
forest = RandomForestRegressor()

# Defining the grid of hyperparameters to search
param_grid = {
    "n_estimators": [50, 100, 150],
    "min_samples_split":[2, 4],
    "max_depth": [None, 4, 8]
    
}

# Creating a GridSearchCV object and fitting it to the scaled training data
# The GridSearchCV object will perform cross-validation for each combination of hyperparameters in the grid and find the best combination
grid_search = GridSearchCV(forest, param_grid, cv=5, 
                           scoring= "neg_mean_squared_error", 
                           return_train_score=True)

grid_search.fit(x_train_s, y_train)
```

```python
# Displaying the best estimator found by the grid search
# The best_estimator_ attribute of a GridSearchCV object is a model with the best combination of hyperparameters

grid_search.best_estimator_
```

```python
# Calculating the R^2 score of the best estimator on the scaled testing data

grid_search.best_estimator_.score(x_test_s, y_test)
```
